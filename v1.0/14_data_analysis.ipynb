```python
t = np.arange(0, 12, 0.1)
# y = np.sin(t)

plt.figure(figsize=(10,6))
plt.title('sin graph', fontsize=20)
plt.plot(t, np.sin(t), label='sin', lw=3, color='purple', linestyle='dashed')
plt.plot(t, np.cos(t), label='cos', lw=3, color='red')

plt.grid() # grid 추가
plt.legend(fontsize=15) # graph에서 legend (graph의 label 출력)
plt.xlabel('time', fontsize=20)
plt.ylabel('amplitude', fontsize=20)
plt.xlim(0, 10) # x의 출력 범위
plt.ylim(-1, 1) # y의 출력 범위
plt.show()
```

```python
t = np.arange(0, 5, 1)

plt.figure(figsize=(10, 5))
plt.plot(t, t, 'r--') # red, dashed
plt.plot(t, t**2, 'bs') # blue, square
plt.plot(t, t**3, 'g^') # green, triangle
plt.xlim(0, 4) # x의 출력 범위
plt.ylim(-1, 10) # y의 출력 범위

plt.show()
```
